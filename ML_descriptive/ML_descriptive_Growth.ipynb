## 2. preprocessing

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import  OneHotEncoder, StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, HistGradientBoostingRegressor
from sklearn.metrics import accuracy_score, r2_score, mean_absolute_percentage_error
```

```python
target = "UNIT_ACCR"

features_list = ['ALT', 'SLOPE25', 'AGE_PPL', 'DIV_STR_PPL', 'DDOM', 'TIGES_VIV_H', 'FEUILL_PER', 'CONIF_PER', 'H_D',
                'NDVI', 'EVI', 'NDMI', 'NDWI', 'DSWI', 'LAST_INTER_DATE', 'PRCP_SUM', 'PRCP', 'PRCP_GROWTH', 'TAVE_AVG',
                'TAVE', 'TAVE_GROWTH', 'PRCP_S_S', 'PRCP_G_S',

                'HT_VEG', 'QUAL_STATION', 'REBOISEMENT_AN', 'TAILLE_PPL',
                
                'LFI', 'ORIENTATION', 'PRODREG', 'UNIT_VEG_FINE', 'UNIT_VEG_GROS', 'MODE_REGEN', 'INTENSITE_EXPLOIT', 'TRACES_FEU', 'NIV_DEV', 'MELANGE',
                'RELIEF', 'STR_PPL', 'TYP_RAJ_PPL', 'ESPECE_DOM', 'FEU_RES', 'TYPE_FORET']
                
# separation of the data
X = work.loc[:, features_list]
Y = work.loc[:, target]

print('Explanatory variables : ', X.columns)

# Divide dataset Train set & Test set 
print("Dividing into train and test sets...")
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
print("...Done.")
```

```python
categorical_features = ['LFI', 'ORIENTATION', 'PRODREG', 'UNIT_VEG_FINE', 'UNIT_VEG_GROS', 'MODE_REGEN', 'INTENSITE_EXPLOIT', 'TRACES_FEU', 'NIV_DEV', 'MELANGE',
                        'RELIEF', 'STR_PPL', 'TYP_RAJ_PPL', 'ESPECE_DOM', 'FEU_RES', 'TYPE_FORET']

ordi_categorical_features = ['HT_VEG', 'QUAL_STATION', 'REBOISEMENT_AN', 'TAILLE_PPL']

numeric_features = ['ALT', 'SLOPE25', 'AGE_PPL', 'DIV_STR_PPL', 'DDOM', 'TIGES_VIV_H', 'FEUILL_PER', 'CONIF_PER', 'H_D',
                    'NDVI', 'EVI', 'NDMI', 'NDWI', 'DSWI', 'LAST_INTER_DATE', 'PRCP_SUM', 'PRCP', 'PRCP_GROWTH', 'TAVE_AVG',
                    'TAVE', 'TAVE_GROWTH', 'PRCP_S_S', 'PRCP_G_S']
                    
# preprocessings
print("Encoding categorical features and standardizing numerical features...")

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(
    steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder', OneHotEncoder(drop='first'))
    ])

ordi_cate_transformer = Pipeline(
    steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('ordi_encoder', OrdinalEncoder(encoded_missing_value=0))
    ])


preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features),
        ('ordicat', ordi_cate_transformer, ordi_categorical_features)
    ])

X_train = preprocessor.fit_transform(X_train)
print("...Done")
print(X_train[0:2,:])

print("Encoding categorical features and standardizing numerical features...")
X_test = preprocessor.transform(X_test)
print("...Done")
print(X_test[0:2,:])
```

## 3. Model linear regression

```python
# Train model
print("Train model...")
model_LR1 = LinearRegression()
model_LR1.fit(X_train, Y_train)
print("...Done.")
```

```python
# predicted values
Y_train_pred = model_LR1.predict(X_train)
Y_test_pred = model_LR1.predict(X_test)

# Print R^2 scores
print("R2 score on training set : ", r2_score(Y_train, Y_train_pred))
print("R2 score on test set : ", r2_score(Y_test, Y_test_pred))

# Print MAPE score
print("MAPE score on training set : ", mean_absolute_percentage_error(Y_train, Y_train_pred))
print("MAPE score on test set : ", mean_absolute_percentage_error(Y_test, Y_test_pred))
```

## 4. Lasso

```python
# Perform 3-fold cross-validation to evaluate the generalized R2 score obtained with a Ridge model
print("10-fold cross-validation...")
model_GL1 = Lasso(alpha=0.01)
scores = cross_val_score(model_GL1, X_train, Y_train, cv=10)
print('The cross-validated R2-score is : ', scores.mean())
print('The standard deviation is : ', scores.std())
```

```python
# Perform grid search
print("Grid search...")
model_GL1 = Lasso()
# Grid of values to be tested
params = {
    'alpha': [0.008, 0.01, 0.03, 0.05]
}
gridsearchL = GridSearchCV(model_GL1, param_grid = params, cv = 10, n_jobs = 6)
gridsearchL.fit(X_train, Y_train)
print("...Done.")
print("Best hyperparameters : ", gridsearchL.best_params_)
print("Best R2 score : ", gridsearchL.best_score_)
```

## 5. Decision Trees

```python
print("DecTree instance...")
model_DT1 = DecisionTreeRegressor(max_depth = 4, min_samples_split = 10)

print("Train model...")
regDT1 = model_DT1.fit(X_train, Y_train)
print("...Done.")
```

```python
# Predictions on training set
Y_train_pred = regDT1.predict(X_train)

# Predictions on test set
Y_test_pred = regDT1.predict(X_test)

# Print scores
print("R2 on training set : ", r2_score(Y_train, Y_train_pred))
print("R2 on test set : ", r2_score(Y_test, Y_test_pred))
```

## 6. Random Forest

```python
# Perform grid search
print("Grid search...")
model_RF1 = RandomForestRegressor()

# Grid of values to be tested
params = {
    'max_depth': [2, 4, 6, 8, 10],
    'min_samples_leaf': [1, 2, 5],
    'min_samples_split': [4, 8, 10, 12],
    'n_estimators': [10, 20, 40, 60, 80, 100]
}
gridsearch = GridSearchCV(model_RF1, param_grid = params, cv = 10, n_jobs = 6)
gridsearch.fit(X_train, Y_train)
print("...Done.")
print("Best hyperparameters : ", gridsearch.best_params_)
print("Best validation accuracy : ", gridsearch.best_score_)
```

```python
# Predictions on training set
print("Predictions on training set...")
Y_train_pred = gridsearch.predict(X_train)
print("...Done.")
print(Y_train_pred)
print()

# Predictions on test set
print("Predictions on test set...")
Y_test_pred = gridsearch.predict(X_test)
print("...Done.")
print(Y_test_pred)
print()
```

```python
# Print scores
print("R2 on training set : ", r2_score(Y_train, Y_train_pred))
print("R2 on test set : ", r2_score(Y_test, Y_test_pred))
print()
```
